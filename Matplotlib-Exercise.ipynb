```python
import numpy as np
x = np.arange(0,100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X variable')
axes.set_ylabel('Y variable')
axes.set_title('Here is a simple chart')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,z,marker='o', markersize=0.5)
ax2.plot(y,z, color='red')
```

```python
plt.show()
```
