```python
df = pd.read_csv('test.csv')
sns.set_style('whitegrid')
```

```python
sns.jointplot(x='profit',y='premium',data=df,kind='kde')
```

```python
sns.pairplot(df,hue='Broker')
```

```python
sns.barplot(x='Broker',y='profit',data=df)
```

```python
sns.countplot(x='Broker',data=df)
```

```python
df
```

```python
df_pivot= df.pivot_table(index='Status',columns='item',values='premium',fill_value=0).astype(int)
df_pivot
```

```python
sns.heatmap(df_pivot,cmap='coolwarm',annot=True,fmt='g')
```

```python
sns.lmplot(x='premium',y='profit',data=df,hue='Broker',markers=['o','v'])
```

```python
sns.lmplot(x='premium',y='profit',data=df,hue='Broker',col='class')
```

```python
sns.countplot(x='class',data=df)
```

```python
plt.figure(figsize=(25,8))
sns.countplot(x='insurer',data=df,palette="hot")
sns.set_context('poster')
```
