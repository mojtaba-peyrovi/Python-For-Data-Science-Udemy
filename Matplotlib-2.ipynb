```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Title')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('first plot')
axes[1].plot(y,x)
axes[1].set_title('second plot')
```

```python
fig = plt.figure(figsize=(8,2))
```

```python
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(5,2))
axes[0].plot(x,y)
plt.tight_layout()
```

```python
fig.savefig('my_plot.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,label='first data')
ax.plot(y,x,label='second data')
ax.legend(loc=0)
```
