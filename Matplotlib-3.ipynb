```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3, alpha=0.5, linestyle='-',marker='o',markersize=10)  # can take rgb hex codes.
ax.set_xlim([0,1])   
ax.set_ylim([0,2])
```
