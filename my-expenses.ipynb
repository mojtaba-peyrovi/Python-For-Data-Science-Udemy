```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('exp.csv')
```

```python
df
```

```python
df.drop(df.index[71:],inplace=True)
```

```python
df.drop(df.columns[6:], axis=1,inplace=True)
```

```python
df
```

```python
df.fillna(value=0)
```

```python
plt.figure(figsize=(20,8))
sns.set_context('poster')
sns.barplot(x='Purpose',y='Spent',data=df)
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='Purpose',data=df)
```

```python
df_pivot= df.pivot_table(index='Date',columns='Purpose',values='Spent',fill_value=0).astype(int)
```

```python
df_pivot
```

```python
sns.heatmap(df_pivot,cmap='coolwarm',fmt='g')
```

```python
plt.figure(figsize=(25,8))
sns.countplot(x='Date',data=df,palette="hot")
sns.set_context('poster')
```
